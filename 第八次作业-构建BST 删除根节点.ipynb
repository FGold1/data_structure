```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def insert_bst(root, val):
    """向 BST 中插入一个值"""
    if root is None:
        return TreeNode(val)
    if val < root.val:
        root.left = insert_bst(root.left, val)
    elif val > root.val:
        root.right = insert_bst(root.right, val)
    return root


def print_tree_visual(root):
    """打印美观的二叉树（精准对齐）"""
    if not root:
        return

    # 收集所有节点和位置
    levels = []
    queue = [(root, 0)]
    while queue:
        node, level = queue.pop(0)
        if len(levels) <= level:
            levels.append([])
        levels[level].append(node)
        if node:
            queue.append((node.left if node.left else None, level + 1))
            queue.append((node.right if node.right else None, level + 1))

    max_depth = len(levels) - 1

    # 计算树的宽度（每层的最大节点数）
    max_width = 2 ** (max_depth + 1)

    # 存储每个节点的显示位置
    pos = {}

    def dfs(node, x, y, width):
        if not node or y > max_depth:
            return
        pos[(x, y)] = str(node.val)
        if node.left:
            dfs(node.left, x - width // 2, y + 1, width // 2)
        if node.right:
            dfs(node.right, x + width // 2, y + 1, width // 2)

    # 初始位置：根节点在中间
    dfs(root, max_width // 2, 0, max_width // 2)

    # 构建树形输出
    result = []

    for y in range(max_depth + 1):
        # 当前层节点行
        node_line = [' '] * (max_width + 1)
        # 下一层的连线行
        line_line = [' '] * (max_width + 1)

        for (x, ly), val in pos.items():
            if ly == y:
                # 放置节点值
                for i, ch in enumerate(val):
                    node_line[x + i] = ch

                # 查找子节点并画连线
                for (cx, cly), cval in pos.items():
                    if cly == y + 1:
                        # 左子节点（在父节点左边）
                        if cx < x and abs(cx + len(cval) // 2 - x) < 5:
                            # 画左斜线
                            line_x = x - 1
                            if line_x >= 0:
                                line_line[line_x] = '/'
                        # 右子节点（在父节点右边）
                        elif cx > x and abs(cx + len(cval) // 2 - (x + len(val) - 1)) < 5:
                            # 画右斜线
                            line_x = x + len(val)
                            if line_x < len(line_line):
                                line_line[line_x] = '\\'

        # 去除尾部空格
        node_str = ''.join(node_line).rstrip()
        line_str = ''.join(line_line).rstrip()

        if node_str:
            result.append(node_str)
        if line_str and y < max_depth:
            result.append(line_str)

    # 输出
    for line in result:
        if line.strip():
            print(line)


# ============================================================
# 构建 BST
# ============================================================
values = [50, 30, 70, 20, 40, 60, 80]
bst_root = None
for v in values:
    bst_root = insert_bst(bst_root, v)

print("=" * 40)
print("构建 BST")
print("=" * 40)
print("插入序列:", values)
print("\n最终 BST 形态：")
print_tree_visual(bst_root)
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def insert_bst(root, val):
    """向 BST 中插入一个值"""
    if root is None:
        return TreeNode(val)
    if val < root.val:
        root.left = insert_bst(root.left, val)
    elif val > root.val:
        root.right = insert_bst(root.right, val)
    return root


def find_max(node):
    """找到子树中的最大值节点（中序前驱）"""
    while node.right:
        node = node.right
    return node


def find_min(node):
    """找到子树中的最小值节点（中序后继）"""
    while node.left:
        node = node.left
    return node


def delete_node_predecessor(root, key):
    """用中序前驱（左子树的最大值）删除节点"""
    if root is None:
        return None
    if key < root.val:
        root.left = delete_node_predecessor(root.left, key)
    elif key > root.val:
        root.right = delete_node_predecessor(root.right, key)
    else:
        # 找到要删除的节点
        if root.left is None:
            return root.right
        if root.right is None:
            return root.left

        # 用中序前驱（左子树的最大值）替换
        predecessor = find_max(root.left)
        root.val = predecessor.val
        # 删除前驱节点
        root.left = delete_node_predecessor(root.left, predecessor.val)
    return root


def delete_node_successor(root, key):
    """用中序后继（右子树的最小值）删除节点"""
    if root is None:
        return None
    if key < root.val:
        root.left = delete_node_successor(root.left, key)
    elif key > root.val:
        root.right = delete_node_successor(root.right, key)
    else:
        # 找到要删除的节点
        if root.left is None:
            return root.right
        if root.right is None:
            return root.left

        # 用中序后继（右子树的最小值）替换
        successor = find_min(root.right)
        root.val = successor.val
        # 删除后继节点
        root.right = delete_node_successor(root.right, successor.val)
    return root


def print_tree_visual(root):
    """打印美观的二叉树（精准对齐）"""
    if not root:
        return

    # 收集所有节点和位置
    levels = []
    queue = [(root, 0)]
    while queue:
        node, level = queue.pop(0)
        if len(levels) <= level:
            levels.append([])
        levels[level].append(node)
        if node:
            queue.append((node.left if node.left else None, level + 1))
            queue.append((node.right if node.right else None, level + 1))

    max_depth = len(levels) - 1

    # 计算树的宽度（每层的最大节点数）
    max_width = 2 ** (max_depth + 1)

    # 存储每个节点的显示位置
    pos = {}

    def dfs(node, x, y, width):
        if not node or y > max_depth:
            return
        pos[(x, y)] = str(node.val)
        if node.left:
            dfs(node.left, x - width // 2, y + 1, width // 2)
        if node.right:
            dfs(node.right, x + width // 2, y + 1, width // 2)

    # 初始位置：根节点在中间
    dfs(root, max_width // 2, 0, max_width // 2)

    # 构建树形输出
    result = []

    for y in range(max_depth + 1):
        # 当前层节点行
        node_line = [' '] * (max_width + 1)
        # 下一层的连线行
        line_line = [' '] * (max_width + 1)

        for (x, ly), val in pos.items():
            if ly == y:
                # 放置节点值
                for i, ch in enumerate(val):
                    node_line[x + i] = ch

                # 查找子节点并画连线
                for (cx, cly), cval in pos.items():
                    if cly == y + 1:
                        # 左子节点（在父节点左边）
                        if cx < x and abs(cx + len(cval) // 2 - x) < 5:
                            # 画左斜线
                            line_x = x - 1
                            if line_x >= 0:
                                line_line[line_x] = '/'
                        # 右子节点（在父节点右边）
                        elif cx > x and abs(cx + len(cval) // 2 - (x + len(val) - 1)) < 5:
                            # 画右斜线
                            line_x = x + len(val)
                            if line_x < len(line_line):
                                line_line[line_x] = '\\'

        # 去除尾部空格
        node_str = ''.join(node_line).rstrip()
        line_str = ''.join(line_line).rstrip()

        if node_str:
            result.append(node_str)
        if line_str and y < max_depth:
            result.append(line_str)

    # 输出
    for line in result:
        if line.strip():
            print(line)


# ============================================================
# 构建 BST
# ============================================================
values = [50, 30, 70, 20, 40, 60, 80]
bst_root = None
for v in values:
    bst_root = insert_bst(bst_root, v)

print("=" * 50)
print("题1：构建 BST")
print("=" * 50)
print("插入序列:", values)
print("\n最终 BST 形态：")
print_tree_visual(bst_root)


# ============================================================
# 题2：删除根节点 50（中序前驱法）
# ============================================================
print("\n" + "=" * 50)
print("题2：删除根节点 50（中序前驱法）")
print("=" * 50)

# 复制原树（重新构建）
root1 = None
for v in values:
    root1 = insert_bst(root1, v)
root1 = delete_node_predecessor(root1, 50)

print("\n删除后的 BST 形态（用 40 替换 50）：")
print_tree_visual(root1)


# ============================================================
# 题2：删除根节点 50（中序后继法）
# ============================================================
print("\n" + "=" * 50)
print("题2：删除根节点 50（中序后继法）")
print("=" * 50)

# 复制原树（重新构建）
root2 = None
for v in values:
    root2 = insert_bst(root2, v)
root2 = delete_node_successor(root2, 50)

print("\n删除后的 BST 形态（用 60 替换 50）：")
print_tree_visual(root2)
```

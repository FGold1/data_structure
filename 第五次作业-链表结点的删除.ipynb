```python
class Node:
    """
    链表结点类。
    每个结点包含：
    - data：存储的数据
    - next_node：指向下一个结点的引用（链）
    """
    def __init__(self, data: str | int | float):
        """
        初始化结点。
        :param data: 结点存储的数据
        """
        self.data: str | int | float = data
        self.next_node: Node | None = None

    def __repr__(self) -> str:
        """返回结点的字符串表示。"""
        return f"Node({self.data!r})"
from typing import List

class LinkedList:
    """
    链表类。
    维护一个指向第一个结点的头指针，
    通过头指针可以遍历整个链表。
    """
    def __init__(self):
        """初始化空链表。"""
        self.first_node: Node | None = None

    def __repr__(self) -> str:
        """返回链表的字符串表示。"""
        if not self.first_node:
            return "LinkedList([])"

        nodes: List[str] = []
        current: Node | None = self.first_node

        while current:
            nodes.append(str(current.data))
            current = current.next_node

        return f"LinkedList([{', '.join(nodes)}])"
def delete_at_index(self, index: int) -> bool:
    """
    删除指定索引位置的元素。
    :param index: 要删除的索引
    :return: 删除成功返回True，索引越界返回False
    """
    if not self.first_node:
        return False

    # 删除开头
    if index == 0:
        self.first_node = self.first_node.next_node
        return True

    # 找到要删除结点的前一个结点
    current_node = self.first_node
    current_index = 0

    while current_index < index - 1:
        if not current_node or not current_node.next_node:
            return False
        current_node = current_node.next_node
        current_index += 1

    if not current_node.next_node:
        return False

    # 删除结点：让前一个结点的next指向要删除结点的下一个
    current_node.next_node = current_node.next_node.next_node
    return True

# 绑定方法
LinkedList.delete_at_index = delete_at_index

# 测试删除
list3 = LinkedList()
list3.first_node = Node("a")
node_b = Node("b")
node_c = Node("c")
node_d = Node("d")
list3.first_node.next_node = node_b
node_b.next_node = node_c
node_c.next_node = node_d

print(f"原链表: {list3}")  # LinkedList([a, b, c, d])

list3.delete_at_index(0)  # 删除开头
print(f"删除索引0后: {list3}")  # LinkedList([b, c, d])

list3.delete_at_index(1)  # 删除索引1（原"c"）
print(f"删除索引1后: {list3}")  # LinkedList([b, d])
```

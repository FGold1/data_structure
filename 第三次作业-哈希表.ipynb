```python
class HashTable:
    def __init__(self,size:int=16):
        """
        初始化哈希表
        :param size: 哈希表大小
        """
        self.size = size
        self.table : list[list[tuple[int,str]]] = [[]for _ in range(size)]
        self.count : int =0
    def _hash(self,key:str)->int:
        """
        哈希函数：将键转化为哈希表的索引
        :param key: 键
        :return: 哈希值
        """
        return hash(key) % self.size
    def _resize(self)->None:
        """
        扩容函数：当负载因子超过阈值时，将哈希表大小扩大一倍
        """
        if self.count < self.size*0.75 :
            return
        old_table=self.table
        self.size *= 2
        self.table = [[]for _ in range(self.size)]
        self.count = 0
        for bucket in old_table:
            for key,value in bucket:
                self.put(value,key)
    def put(self,key:str,value:int)->None:
        """
        向哈希表中插入或更新键值对
        :param key: 键
        :param value: 值
        """
        #计算哈希值
        index = self._hash(key)
        bucket = self.table[index]

        for old_index,(old_value,old_key) in enumerate(bucket):
            if key==old_key:
                bucket[old_index] = (value,key)
                return
        bucket.append((value,key))
        self.count += 1
        self._resize()
    def get(self,key:str)->int|None:
        """
        获取键的对应值
        :param key: 键
        :return: 值，不存在则返回None
        """
        index = self._hash(key)
        bucket = self.table[index]
        for old_value,old_key in enumerate(bucket):
            if old_key==key:
                return old_value
        return None
    def delete(self,key:str)->bool:
        """
        删除键值对
        :param key: 键
        :return: 删除成功返回True，键不存在返回False
        """
        index = self._hash(key)
        bucket = self.table[index]
        for old_index,(old_key,old_value) in enumerate(bucket):
            if old_key==key:
                del bucket[old_index]
                self.count-=1
                return True
        return False
    def _str_(self)->str:
        """
        返回哈希表的字符串表示
        :return: 哈希表的字符串表示
        """
        items ={}
        for bucket in self.table:
            for key,value in bucket:
                items[key]=value
        return str(items)

ht = HashTable()
ht.put("apple", 100)
ht.put("banana", 200)
ht.put("orange", 300)

result = ht.delete("banana")
print(f"删除结果: {result}")

```

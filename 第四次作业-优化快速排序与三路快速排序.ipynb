```python
import random
from collections.abc import Sequence
from typing import Optional


class RandomizedQuickSort:
    """随机化快速排序，通过随机选择轴来避免最坏情况。"""

    def __init__(self, arr: Sequence[int | float]):
        self.arr: list[int | float] = list(arr)
        self.length: int = len(arr)

    def partition(self, left_index: int, right_index: int) -> int:
        """分区操作，使用随机选择的轴。"""
        # 随机选择轴，并与最右边元素交换
        random_pivot = random.randint(left_index, right_index)
        self.arr[random_pivot], self.arr[right_index] = self.arr[right_index],self.arr[random_pivot]

        pivot: int | float = self.arr[right_index]
        left_pointer: int = left_index
        right_pointer: int = right_index - 1

        while True:
            # 左指针向右移动，直到找到大于或等于轴的值
            while left_pointer <= right_pointer and self.arr[left_pointer] < pivot:
                left_pointer += 1
            # 右指针向左移动，直到找到小于或等于轴的值
            while left_pointer <= right_pointer and self.arr[right_pointer] > pivot:
                right_pointer -= 1

            # 如果左右指针相遇或交叉，停止
            if left_pointer >= right_pointer:
                break

            # 交换左右指针指向的值
            self.arr[left_pointer], self.arr[right_pointer] = (
                self.arr[right_pointer],
                self.arr[left_pointer],
            )
            # 交换后移动指针，避免死循环
            left_pointer += 1
            right_pointer -= 1

        # 最后将左指针的值与轴交换
        self.arr[left_pointer], self.arr[right_index] = (
            self.arr[right_index],
            self.arr[left_pointer],
        )
        return left_pointer

    def sort(self,left_index: Optional[int] = None,right_index: Optional[int] = None) -> list[int |float]:
        """递归进行快速排序。"""
        if left_index is None:
            left_index = 0
        if right_index is None:
            right_index = self.length - 1

        if right_index - left_index <= 0:
            return self.arr

        pivot_position = self.partition(left_index, right_index)
        self.sort(left_index, pivot_position - 1)
        self.sort(pivot_position + 1, right_index)

        return self.arr


# 测试随机化快速排序

test_arr = [64, 34, 25, 12, 22, 11, 90, 5, 5, 5, 77, 33]
print(f"排序前: {test_arr}")
rqs = RandomizedQuickSort(test_arr)
print(f"排序后: {rqs.sort()}")
```

```python
from collections.abc import Sequence
from typing import Optional

class ThreeWayQuickSort:
    """三路快速排序，适用于含有大量重复元素的数组。"""

    def __init__(self, arr: Sequence[int | float]):
        self.arr: list[int | float] = list(arr)
        self.length: int = len(arr)

    def sort(self,left_index: Optional[int] = None,right_index: Optional[int] = None) -> list[int |float]:
        if left_index is None:
            left_index = 0
        if right_index is None:
            right_index = self.length - 1

        if right_index <= left_index:
            return self.arr

        # 三路分区
        lt, gt = self.three_way_partition(left_index, right_index)

        # 递归排序小于轴和大于轴的部分
        self.sort(left_index, lt - 1)
        self.sort(gt + 1, right_index)

        return self.arr

    def three_way_partition(self, left: int, right: int) -> tuple[int, int]:
        """
        三路分区：将数组分为 [小于轴] [等于轴] [大于轴] 三部分。

        :return: (lt, gt) 其中：
                 lt: 等于轴部分的左边界
                 gt: 等于轴部分的右边界
        """
        pivot: int | float = self.arr[left]
        lt: int = left      # arr[left+1 ... lt] < pivot
        i: int = left + 1   # arr[lt+1 ... i-1] == pivot
        gt: int = right     # arr[gt ... right] > pivot

        while i <= gt:
            if self.arr[i] < pivot:
                self.arr[lt], self.arr[i] = self.arr[i], self.arr[lt]
                lt += 1
                i += 1
            elif self.arr[i] > pivot:
                self.arr[i], self.arr[gt] = self.arr[gt], self.arr[i]
                gt -= 1
            else:  # self.arr[i] == pivot
                i += 1

        return lt, gt

test_arr2 = [5, 2, 5, 1, 5, 3, 5, 4, 5, 5, 5, 2, 2, 3, 3, 4, 4, 1, 1]
print(f"排序前: {test_arr2}")
twqs = ThreeWayQuickSort(test_arr2)
print(f"排序后: {twqs.sort()}")



```

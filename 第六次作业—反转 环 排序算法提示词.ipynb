```python
class DNode:
    """双向链表结点"""
    def __init__(self, data: str | int | float):
        self.data = data
        self.prev: DNode | None = None
        self.next: DNode | None = None

class DoublyLinkedList:
    """双向链表"""
    def __init__(self):
        self.first: DNode | None = None
        self.last: DNode | None = None

    def insert_at_end(self, value: str | int | float) -> None:
        """在末尾插入 - O(1)"""
        new_node = DNode(value)
        if not self.first:  # 空链表
            self.first = self.last = new_node
        else:
            # 新节点的prev指向原来的队尾
            new_node.prev = self.last
            # 原队尾的next指向新节点
            self.last.next = new_node
            # 新队尾指向新节点
            self.last = new_node

    def remove_from_front(self) -> str | int | float | None:
        """从开头删除 - O(1)"""
        if not self.first:
            return None
        data = self.first.data
        self.first = self.first.next
        if self.first:
            self.first.prev = None
        else:
            self.last = None
        return data
    def reverse(self) -> None:
        """反转双向链表"""
        if not self.first:  # 空链表
            return
        current = self.first
    # 交换每个结点的prev和next
        while current:
            current.prev, current.next = current.next, current.prev
            current = current.prev  # 注意：交换后原next变成了prev
    # 交换first和last
        self.first, self.last = self.last, self.first
    def has_cycle(self) -> bool:
        """检测链表中是否存在环（快慢指针法）"""
        if not self.first:
            return False

        slow = self.first
        fast = self.first

        while fast and fast.next:
            slow = slow.next          # 慢指针每次走一步
            fast = fast.next.next     # 快指针每次走两步
            if slow == fast:          # 相遇说明有环
                return True

        return False
```

```python
# 测试反转
dll = DoublyLinkedList()
for v in ['a', 'b', 'c']:
    dll.insert_at_end(v)

# 打印原链表
current = dll.first
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")

dll.reverse()  # 反转

# 打印反转后的链表
current = dll.first
while current:
    print(current.data, end=" -> ")
    current = current.next
print("None")


# 测试环检测
print(f"正常链表是否有环: {dll.has_cycle()}")  # False

# 手动构造一个环（仅供测试）
dll.last.next = dll.first  # 尾指向头，形成环
print(f"构造环后是否有环: {dll.has_cycle()}")  # True
```
